### Set Up The Data For The Pie Chart

```python
# Unique function
def unique(list1):
    unique_list = []
    for x in list1:
        # check if exists in unique_list or not
        if x not in unique_list:
            unique_list.append(x)
    return unique_list
```

```python
# Frequency of different teams
unique_season_years = unique(player_stats['season year'])
print(unique_season_years)
player_stats.loc[player_stats['season year'] == '13/14']['team']
pie_array = []
for i in unique(player_stats['team']):
    pie_array.append([i, 0])
for i in unique_season_years:
    #print(i)
    for j in player_stats.loc[player_stats['season year'] == i]['team'].unique().tolist():
        #print(j)
        for k in pie_array:
            if k[0] == j:
                k[1] += 1
                break
print(pie_array)
```

```python
# Frequency of how many season they played in the champion league
season_played_in_league_label = [1, 2, 3, 4, 5, 6, 7, 8]
season_played_in_league_data = [0, 0, 0, 0, 0, 0, 0, 0]

for i in pie_array:
    for j in range(1,9):
        #print(j)
        #print(pie_array[1])
        if i[1] == j:
            #print("HIOL")
            season_played_in_league_data[j-1] += 1
            break

print(season_played_in_league_data)
```

## Question: Which Team Has Gotten The Most Yellow And Red Cards.

### Create an appropriate pandas series
* REMINDER: Use .groupby() to organize your series based on another column's data

```python
# Frequency of yellow cards for each team
#all_cards = player_stats.groupby(['team'])["yellowCards", "directRedCards"].apply(lambda x : x.astype(int).sum())
yellow_cards = player_stats.groupby(['team'])['yellowCards'].sum()
red_cards = player_stats.groupby(['team'])['directRedCards'].sum()
all_cards = yellow_cards + red_cards
all_cards[:10]
```

### Split into individual lists
* Create two parallel lists 
    * One list of categories
    * One list of statistics of interest
* Show each list individually
    * This is a good way for you to see if your modifications were successful and to see what the data looks like

```python
# Create team name list
team_names_card = all_cards.index.tolist()
# Create frequency list
team_count_card = all_cards.tolist()
# Print
print(team_names_card[:10])
print(team_count_card[:10])
```

### Create list of sorted tuples
* Loop through parallel lists to create tuples in the form (category, statistic)
* Sort appropriately
    * REMINDER: Use list_name.sort(key=lambda x: x[1], reverse=True) to sort
* Show the list
    * This is a good way for you to see if your modifications were successful and to see what the data looks like

```python
# Turn my categorti and stats into a tuple
all_cards_tuple = []
for i in range(len(team_names_card)):
    all_cards_tuple.append((team_names_card[i], team_count_card[i]))
all_cards_tuple[:10]
```

### Split into individual sorted lists
* Option #1

### Create a 'bar position list'

```python
# Frequency of yellow cards for each team
player_goal_count = player_stats.groupby(['name', 'goals'])['goals'].sum()
player_goal_count = player_goal_count.index.tolist()
player_goal_count[:10]
```

## Set graph details
* Import matplotlib appropriately
* Your graph should be styled uniquely compared to any previously created, and must inlude:
    * Title
    * X-axis label
    * Y-axis label
    * Bar labels (xticks)
    * Visually appealing style
    * Graph color not previously used

```python
# Import matplotlib and numpy
import matplotlib.pyplot as plot
%matplotlib inline

# Graph details
plot.style.use("ggplot")
plot.figure(figsize=(30,5))
plot.title('Total Goals By Each Player (Only 250 Players Included)')
plot.ylabel('Goals Scored')
plot.xlabel('Name')

x_data = []
y_data = []
for i in player_goal_count:
    x_data.append(i[0])
    y_data.append(i[1])
print(x_data[:10])
print(y_data[:10])

length = len(player_goal_count)
bar_pos = range(0, 30)
bar_pos = np.arange(1, 251)
print(bar_pos)
width = 0.9

plot.bar(x_data[:250], y_data[:250], width, color="tan")
#print(bar_pos)
plot.xticks(bar_pos, x_data[:250], rotation=90)
# Bar graph
plot.show()
```
